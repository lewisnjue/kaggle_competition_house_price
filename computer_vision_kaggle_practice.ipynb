```python
# create a tranformation for our images
import torchvision.transforms as T
image_transform = T.Compose([
    T.Resize((64,64)),
    T.ToTensor()
])
```

```python
# import the images using Image forlder
from torchvision.datasets import ImageFolder
image_foder = pathlib.Path('images/animals/')
image_data = ImageFolder(image_foder, transform=image_transform)
```
